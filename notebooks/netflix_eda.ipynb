```python
import pandas as pd

df = pd.read_csv('../data/Netflix_Movies_and_TV_Shows.csv')
df.head()
```

```python
df.shape
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

Netflixâ€™s catalog contains more TV shows than movies, showing a stronger emphasis on content with episodes.

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
movies_by_year = df['release_year'].value_counts().sort_index()

movies_by_year.tail(10)
```

Most titles in the dataset were released after 2014, showing a significant increase in Netflix content in recent years

```python
df['release_year'].value_counts().sort_index().plot(figsize=(10,5))
plt.title('Netflix Content Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.show()
```

```python
df['genre'].value_counts().head(10)
```

The dataset contains eight unique genres, with Horror and Thriller being the most common, indicating a preference for suspense-driven content.

```python
df['genre'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title('Top Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.show()
```

```python
df['country'].value_counts().head(10)
```

Netflix content is produced across multiple countries, with Canada and the United Kingdom contributing the highest number of titles.

```python
df['country'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title('Top Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```

## Conclusion

This exploratory data analysis provided insights into Netflixâ€™s content distribution across type, genre, release year, and country. The platform shows a strong emphasis on TV shows, recent releases, and popular genres such as Horror and Thriller. It also shows top content producing regions across the world.
